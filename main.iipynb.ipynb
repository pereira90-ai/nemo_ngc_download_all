```python

all_models = {}
for mdo_name in nlp_base_models:
    # Get the attribute itself  
    mdo = getattr(nemo_nlp.models, mdo_name)  
    
    # Check if the attribute is callable and has the 'play' method  
    if callable(mdo) and hasattr(mdo, 'list_available_models'):  
        play_method = getattr(mdo, 'list_available_models')  
        
        # Ensure 'play' is callable before calling it  
        if callable(play_method):  
            print(mdo_name)
            all_models[mdo_name] = play_method()

            if play_method() is not None:
                for one_model in play_method():
                    print(one_model.pretrained_model_name)
                    # try:
                    #     model = mdo.from_pretrained(one_model.pretrained_model_name)
                    #     if not os.path.exists(mdo_name):
                    #         os.mkdir(mdo_name)
                    #     model.save_to(mdo_name + '/' + one_model.pretrained_model_name + '.nemo')
                    # except Exception as e:
                    #     print('error:' , e)
```
